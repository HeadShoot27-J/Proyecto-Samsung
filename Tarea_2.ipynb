```python
import pandas as pd
```

```python

df=pd.read_csv(r"EncuestaPA_JV.csv")
New_df = df.set_index('email')
print(New_df.info())
New_df.head(len(New_df))
```

```python
def reemplazar_valores_multiple_columnas(df, columnas, reemplazo_si, reemplazo_no):
  for col in columnas:
    if col not in df.columns:
      print(f"La columna '{col}' no existe en el DataFrame.")
      continue
    df[col] = df[col].replace({'Sí': reemplazo_si, 'No': reemplazo_no})

 
  df = df.dropna(subset=columnas)
```

```python
print(New_df.columns)

columnas_a_modificar = ['bool_ahorro', 'educacion_financiera', 'bool_educacion_financiera ']

reemplazar_valores_multiple_columnas(New_df, columnas_a_modificar, 1, 0)

New_df.head(len(New_df))
```
