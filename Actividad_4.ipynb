```python
import numpy as np
import pandas as pd

mi_dataframe = pd.read_csv(r"EncuestaPA_JV.csv", index_col='email')
#mi_dataframe.head(len(mi_dataframe))
#print(mi_dataframe.columns.tolist())
```

```python
def filtrar_dataframe(df, columna, valores_a_eliminar, indice='email'):
    mascara = ~df[columna].isin(valores_a_eliminar)
    df_filtrado = df[mascara]
    return df_filtrado
```

```python
valores_a_eliminar = ['No aplica.', np.nan, 'no tengo actividades extracurriculares']
columna_a_filtrar = 'act_extracurricular'

df_filtrado = filtrar_dataframe(mi_dataframe, columna_a_filtrar, valores_a_eliminar)

df_filtrado.head(len(df_filtrado))
```
